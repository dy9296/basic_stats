```python
import numpy as np
import matplotlib.pyplot as plt

# plot functions
def plot_pdf(variable, pdf):
    plt.plot(variable, pdf)
    plt.title("prob. density function")
    plt.xlabel("$x$")
    plt.ylabel("$p(x)$")
    plt.show()

def plot_cdf(variable, cdf):
    plt.plot(variable, cdf)
    plt.title("cumm. density function")
    plt.xlabel("$x$")
    plt.ylabel("$p(x)$")
    plt.show()

def cdf_histogram(cdf):
    n_data = cdf.shape[0]
    
    cls = np.digitize(np.random.uniform(0, 1, n_data * 10), bins = cdf)
    freq = np.unique(cls, return_counts=True)

    plt.bar(freq[0], freq[1], width=1.0)
    plt.title("randaom var. histogram")
    plt.show()
```

```python
# normal distribution

from distribution import normal as gaussian_dist
from scipy.stats import norm as gauss_ans

# normal plot
gaussian = gaussian_dist(2, 2)
var_x = np.linspace(-10, 10, 100)

# gamma pdf
gauss_pdf = gaussian.pdf(var_x)
print(gauss_pdf.round(6) == gauss_ans(2, np.sqrt(2)).pdf(var_x).round(6))
plot_pdf(var_x, gauss_pdf)


# gamma cdf
gauss_cdf = gaussian.cdf(var_x)
plot_cdf(var_x, gauss_cdf)
plt.show()

# gamma R.V. hist
cdf_histogram(gauss_cdf)
```

```python
# bernoulli distribution

from distribution import bernoulli as bern_dist
from scipy.stats import bernoulli

# normal pdf
var_x = np.array([0, 1])

result = bern_dist(0.7).pdf(var_x)
answer = bernoulli(0.7).pmf(var_x)

print(result.round(6) == answer.round(6))
plot_pdf(var_x, result)
```

```python
# gamma distribuiton

from distribution import gamma as gamma_dist

from scipy.special import gamma as gamma_fun
from scipy.stats import gamma as gamma_ans

# gamma function
gamma = gamma_dist(5)
print(gamma.function(True).round(6), gamma_fun(5).round(6))

# gamma plot
var_x = np.linspace(1, 20, 100)

# gamma pdf
gamma_pdf = gamma.pdf(var_x)
print(gamma_pdf.round(6) == gamma_ans(5).pdf(var_x).round(6))
plot_pdf(var_x, gamma_pdf)

# gamma cdf
gamma_cdf = gamma.cdf(var_x)
plot_cdf(var_x, gamma_cdf)
plt.show()

# gamma R.V. hist
cdf_histogram(gamma_cdf)
```

```python
# beta distribuiton

from distribution import beta as beta_dist

from scipy.special import beta as beta_fun
from scipy.stats import beta as beta_ans

# beta function
beta = beta_dist(0.5, 0.5)
print(beta.function().round(6), beta_fun(0.5, 0.5).round(6))

# beta plot
var_x = np.linspace(0.01, 0.99, 100)

# beta pdf
beta_pdf = beta.pdf(var_x)
print(beta_pdf.round(6) == beta_ans(0.5, 0.5).pdf(var_x).round(6))
plot_pdf(var_x, beta_pdf)
```
