```python
import numpy as np
import matplotlib.pyplot as plt

def plot_pdf(variable, pdf):
    plt.plot(var_x, result)
    plt.title("prob. density function")
    plt.xlabel("$x$")
    plt.ylabel("$p(x)$")
    plt.show()
```

```python
# normal distribution

from distribution import normal as gaussian_dist
from scipy.stats import norm

# normal pdf
var_x = np.linspace(-5, 5, 100)

result = gaussian_dist(0, 2).pdf(var_x)
answer = norm(0, np.sqrt(2)).pdf(var_x)

print(result.round(6) == answer.round(6))
plot_pdf(var_x, result)
```

```python
# gamma distribuiton

from distribution import gamma as gamma_dist

from scipy.special import gamma as gamma_fun
from scipy.stats import gamma as gamma_ans

# gamma function
print(gamma_dist(0.5).function(True).round(6), gamma_fun(0.5).round(6))

# gamma pdf
var_x = np.linspace(1, 20, 100)

result = gamma_dist(5).pdf(var_x)
answer = gamma_ans(5).pdf(var_x)

print(result.round(6) == answer.round(6))
plot_pdf(var_x, result)
```

```python
# beta distribuiton

from distribution import beta as beta_dist

from scipy.special import beta as beta_fun
from scipy.stats import beta as beta_ans

# beta function
print(beta_dist(0.5, 0.5).function().round(6), beta_fun(0.5, 0.5).round(6))

# beta pdf
var_x = np.linspace(0.01, 0.99, 100)

result = beta_dist(0.5, 0.5).pdf(var_x)
answer = beta_ans(0.5, 0.5).pdf(var_x)

print(result.round(6) == answer.round(6))
plot_pdf(var_x, result)
```
